# Building the Mobile Application
```
git clone https://github.com/NeuralNet-Hub/yolo-flutter-app.git
cd yolo-flutter-app
docker build -t yolo-flutter-app-flutter-builder .
docker run -it -v $(pwd):/app/ yolo-flutter-app-flutter-builder:latest /bin/bash
```

```
cd example
flutter build apk --release
```

# Exporting Models for TensorRT
```
yolo export format=engine imgsz=384,640 half=True simplify=True device=0 batch=1 model=yolo26s.pt

yolo export format=engine imgsz=384,640 half=True simplify=True device=0 batch=1 model=yolo26x.pt
```

# Running Predictions

```
yolo predict detect device=0 source=video_example/ model=yolo26s_1x384x640_half.engine half=True imgsz=384,640

yolo predict detect device=0 source=video_example/ model=yolo26s.pt
```
