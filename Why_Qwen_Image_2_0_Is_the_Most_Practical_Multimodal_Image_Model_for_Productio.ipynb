# Editing with a mask

```python
import base64
import requests

API_URL = "https://api.your-qwen-provider.com/v1/image/generate"
API_KEY = "YOUR_API_KEY"
def encode_image(path):
    with open(path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")
input_image = encode_image("original_poster.png")
mask_image = encode_image("mask_title_area.png")  # white = editable, black = keep
prompt = """
Replace the title text with: 'Qwen-Image 2.0 Launch'.
Keep style and colors consistent with the original design.
"""
payload = {
    "prompt": prompt,
    "mode": "edit",
    "image": input_image,
    "mask": mask_image,
    "width": 2048,
    "height": 2048,
    "seed": 12345
}
headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}
resp = requests.post(API_URL, json=payload, headers=headers)
resp.raise_for_status()
edited_image_base64 = resp.json()["images"][0]
```
