# Fine-tuning

```python
from rfdetr import RFDETRBase

model = RFDETRBase()
model.train(
    dataset_dir="path/to/dataset",
    epochs=20,  # Yes, 20 is often enough
    batch_size=4,
    lr=1e-4,
    output_dir="checkpoints/",
)
```

# TensorRT Export

```python
model.export(format="tensorrt", fp16=True)
```
